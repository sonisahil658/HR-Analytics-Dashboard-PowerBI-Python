```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Documents\hr_analytics.csv.csv")
```

```python
print(df.shape)
```

```python
print(df.head())
```

```python
cols_to_drop = ['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
                'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19']
df = df.drop(columns=cols_to_drop)
print(df.shape)  # should now show 22214 rows, 14 columns
```

```python
df = df.dropna(subset=['first_name'])
print(df.shape)  # should now show 22213 rows
```

```python
df['birthdate'] = pd.to_datetime(df['birthdate'], format='%d-%m-%Y')
print(df['birthdate'].dtype)   # should say datetime64
print(df['birthdate'].head(3))
```

```python
df['hire_date'] = pd.to_datetime(df['hire_date'], format='%d-%m-%Y')
print(df['hire_date'].dtype)
print(df['hire_date'].head(3))
```

```python
df['termdate'] = pd.to_datetime(df['termdate'], utc=True, errors='coerce')
df['termdate'] = df['termdate'].dt.date
print(df['termdate'].head(5))
print(df['termdate'].isnull().sum())  # should still show ~18285 nulls — this is correct
```

```python
df['Age'] = df['Age'].astype(int)
print(df['Age'].dtype)   # should say int64
print(df['Age'].describe())
```

```python
print(df.dtypes)
print(df.isnull().sum())
```

```python
df = df.reset_index(drop=True)
print(df.shape)
print(df.head(3))
```

```python
df.to_csv("hr_analytics_clean.csv", index=False)
print("Clean file saved successfully.")
```
